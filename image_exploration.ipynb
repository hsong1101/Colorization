Only using test images of datasset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import subprocess
```

```python
def extract_images(path):
    
    global i
    
    for file in os.listdir(path):
        
        if os.path.isdir(path + '/' + file):
            extract_images(path+'/'+file)

        else:
            if '.jpg' in file:
    
                try:
                    img = plt.imread(path+'/'+file)
                    img = cv2.resize(img, (224, 224))

                    plt.imsave(f'data/images/{i}.jpg', img)
                except:
                    continue
                    
                print(f'{i}', flush=True, end='\r')
                
                i += 1
```
